```python
# use: brew install graphviz
from sklearn import tree
#features = [[140,1], [130,1], [150,0], [170, 0],[90,0]]
features = [[130,0], [140, 0], 
            [150, 1], [160, 1],
           [80, 1]]
labels = [0,0,1,1,2]
target_names = ['Apple', 'orange', 'mandarin']
features_name = ['weight', 'texture']
clf = tree.DecisionTreeClassifier()
clf = clf.fit(features, labels)
#print(clf.predict([[160,0]]))
tree.plot_tree(clf.fit(features, labels))
```

```python
# Generating visualization
from sklearn.externals.six import StringIO
import pydot
dot_data = StringIO()
tree.export_graphviz(clf, out_file=dot_data,
 feature_names=features_name,
 class_names=target_names,
 filled=True, rounded=True,
 impurity=False)
graph = pydot.graph_from_dot_data(dot_data.getvalue())
graph[0].write_pdf('fruit.pdf')
!open fruit.pdf
```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
test_idx = [0, 50, 100]

# training data
train_target = np.delete(iris.target, test_idx)
train_data = np.delete(iris.data, test_idx, axis=0)

# testing data
test_target = iris.target[test_idx]
test_data = iris.data[test_idx]

clf = tree.DecisionTreeClassifier()
clf.fit(train_data, train_target)

print(test_target)
print(clf.predict(test_data))

# viz code
from sklearn.externals.six import StringIO
import pydot
dot_data = StringIO()
tree.export_graphviz(clf,
        out_file=dot_data,
        feature_names=iris.feature_names,
        class_names=iris.target_names,
        filled=True, rounded=True,
        impurity=False)

graph = pydot.graph_from_dot_data(dot_data.getvalue())
graph[0].write_pdf("iris.pdf")
!open iris.pdf
```
