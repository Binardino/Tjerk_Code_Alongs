```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
iris_df['target'] = iris.target
```

```python
iris_df
```

```python
iris_df['sepal length (cm)'].plot(kind='bar')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(20, 10))
ax[0, 1].bar(iris_df.target, iris_df['sepal length (cm)'])
ax[0, 1].set_title('hello')
ax[0, 1].set_xlabel('flower species')
ax[0, 1].set_ylim(0, 10)
ax[0, 0].bar(iris_df.index, iris_df['sepal length (cm)'])
```

```python
sns.barplot(iris_df.target, iris_df['sepal length (cm)'])
```

```python
plt.subplots( figsize=(20,10))
x = sns.countplot(x=iris_df['sepal length (cm)'], color='blue')
x.set_title('Sepal length count')
```

```python
plt.subplots( figsize=(20,10))
plt.bar(sorted(iris_df['sepal length (cm)'].unique()), iris_df['sepal length (cm)'].value_counts().sort_index())
```

```python
sorted(iris_df['sepal length (cm)'].unique())
```

```python
iris_df['sepal length (cm)'].value_counts().sort_index()
```

```python
sns.pairplot(iris_df, hue='target')
```

```python
fig, ax = plt.subplots()
sns.lineplot(iris_df['sepal length (cm)'].index, iris_df['sepal length (cm)'], ax=ax, legend='brief')
sns.lineplot(iris_df['sepal width (cm)'].index, iris_df['sepal width (cm)'], ax=ax, legend='brief')
ax.legend(['length', 'width'])
ax.set_xticklabels([1,2,3,4,5])
ax.set_ylabel('cm')
```
